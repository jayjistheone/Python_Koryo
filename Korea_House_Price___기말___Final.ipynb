```python
import pandas as pd
import numpy as np
after_graduation=pd.read_excel("/content/drive/MyDrive/2021_연도별 고등학교 세부 학제별 졸업 후 상황.xlsx",header=3)
# KESS 교육통계서비스(https://kess.kedi.re.kr/index) 에서 고등학교 졸업후 상황 자료를 바탕으로
# 일명 강남8학군 지역의 학생들의 진학률 및 취업자 수를 타 지역과 비교하고,
# 실제 8학군 지역의 학생들이 높은 진학률과 높은 취업자 수를 나타내는지 확인해 보고
# 교육열의 정도가 진학 및 취업에 연관이 있는지를 파악 하고
# 집값에도 영향이 있는지 확인하려 한다.
after_graduation

name=after_graduation
after_graduation.rename(columns={name.columns[0]:'Year',
                         name.columns[1]:'City',
                         name.columns[2]:'High_School',
                         name.columns[3]:'HighSchool_Graduate_Overall',
                         name.columns[4]:'HighSchool_Graduate_Female',
                         name.columns[5]:'Newly_Enrolled_Overall',
                         name.columns[6]:'Newly_Enrolled_Female',
                         name.columns[11]:'4yr_University_Enrolled_Overall',
                         name.columns[12]:'4yr_University_Enrolled_Female',
                         name.columns[19]:'Newly_Employeed_Overall',
                         name.columns[20]:'Newly_Employeed_Female'}, inplace=True)
after_graduation['HighSchool_Graduate_Male']=after_graduation['HighSchool_Graduate_Overall'] -after_graduation['HighSchool_Graduate_Female']
after_graduation['Newly_Enrolled_Male']=after_graduation['Newly_Enrolled_Overall'] -after_graduation['Newly_Enrolled_Female']
after_graduation['4yr_University_Enrolled_Male']=after_graduation['4yr_University_Enrolled_Overall'] -after_graduation['4yr_University_Enrolled_Female']
after_graduation['Newly_Employeed_Male']=after_graduation['Newly_Employeed_Overall'] -after_graduation['Newly_Employeed_Female']

after_graduation['Uni_Enrol_Rate']=after_graduation['Newly_Enrolled_Overall'] /after_graduation['HighSchool_Graduate_Overall']
after_graduation['4yr_Uni_Enrol_Rate']=after_graduation['4yr_University_Enrolled_Overall'] /after_graduation['HighSchool_Graduate_Overall']


after_graduation=after_graduation.drop(['전체.2','여자.2','전체.3','여자.3','전체.5','여자.5','전체.6','여자.6','전체.7','여자.7','전체.9','전체.10','여자.9'],axis=1)
after_graduation=after_graduation.iloc[:2448]
after_graduation
```
